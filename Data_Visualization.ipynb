## Q3. Comapring Different Categories

```python
data = {
    'Category' : ['A','B','C','D','E']*10,
    'Value' : np.random.randint(1,100,50)
        }

df = pd.DataFrame(data)
```

```python
df.head(10)
```

```python
category_means = df.groupby('Category')['Value'].mean().reset_index()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Category',y='Value',data=category_means)
plt.title('Comparison of Different Categories')
plt.xlabel('Category')
plt.ylabel('Mean Value')
plt.show()
```

## Q5. Heat Map

```python
np.random.seed(0)
n_samples = 1000
n_features = 5

corr_matrix = np.array([
    [1.0,0.8,0.4,0.2,0.1],
    [0.8,1.0,0.7,0.5,0.3],
    [0.4,0.7,1.0,0.9,0.6],
    [0.2,0.5,0.9,1.0,0.8],
    [0.1,0.3,0.6,0.8,1.0]
])

data = np.random.multivariate_normal(mean=[0,0,0,0,0],cov=corr_matrix,size=n_samples)

df = pd.DataFrame(data,columns=['Feature 1','Feature 2','Feature 3','Feature 4','Feature 5'])
```

```python
df.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title('Heat Map')
plt.show()
```

# PLOTLY Assignment

## Q1. Scatter Plot

```python
import plotly.graph_objects as go

np.random.seed(30)

data = {
    'X':np.random.uniform(-10,10,300),
    'Y':np.random.uniform(-10,10,300),
    'Z':np.random.uniform(-10,10,300)
}

df = pd.DataFrame(data)
```

```python
fig = go.Figure(data=[go.Scatter3d(x=df['X'],y=df['Y'],z=df['Z'],mode='markers', marker=dict(size=5,color=np.random.rand(300),opacity=0.8))])

fig.update_layout(scene=dict(xaxis_title='X',yaxis_title='Y',zaxis_title='Z',xaxis=dict(nticks=5),yaxis=dict(nticks=5),zaxis=dict(nticks=5)),title='3D Scatter Plot',width=800,height=800)

fig.show()
```

## Q2. Violin Plot

```python
np.random.seed(15)
data = {
    'Grade':np.random.choice(['A','B','C','D','E'],200),
    'Score':np.random.randint(50,100,200)
}

df = pd.DataFrame(data)
```

```python
fig = go.Figure(data=[go.Violin(x=df['Grade'],y=df['Score'],name='Violin Plot',box_visible=True,line_color='blue',fillcolor='lightblue',meanline_visible=True)])

fig.update_layout(title='Violin Plot Showing Distribution of Grades',xaxis_title='Grade',yaxis_title='Score',width=800,height=800)

fig.show()
```

## Q3. Heat Map

```python
np.random.seed(20)
data = {
    'Month':np.random.choice(['Jan','Feb','Mar','Apr','May'],100),
    'Day':np.random.choice(range(1,31),100),
    'Sales':np.random.randint(1000,5000,100)
}

df = pd.DataFrame(data)
```

```python
df_pivot = df.pivot_table(index='Month',columns='Day',values='Sales',aggfunc='mean')
```

```python
fig = go.Figure(data=go.Heatmap(z=df_pivot.values,x=df_pivot.columns,y=df_pivot.index,colorscale='Viridis',reversescale=True))

fig.update_layout(title='Heat Map Showing Monthly Sales',xaxis_title='Day',yaxis_title='Month')

fig.show()
```

## Q4. 3D surface for a function

```python
x= np.linspace(-5,5,100)
y= np.linspace(-5,5,100)
x,y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X':x.flatten(),
    'Y':y.flatten(),
    'Z':z.flatten()
}
df = pd.DataFrame(data)
```

```python
fig = go.Figure(data=[go.Surface(x=x,y=y,z=z,colorscale='Viridis',reversescale=True)])

fig.update_layout(title='3D Surface for a Function',scene=dict(xaxis_title='X',yaxis_title='Y',zaxis_title='Z'),width=800,height=800)

fig.show()
```

## Q5. Bubble Chart

```python
np.random.seed(25)
data = {
    'Country':['USA','Canada','UK','Germany','France'],
    'Population':np.random.randint(100,1000,5),
    'GDP':np.random.randint(500,2000,5)
}
df = pd.DataFrame(data)
```

```python
fig = go.Figure(data=[go.Scatter(x=df['GDP'],y=df['Population'],mode='markers',text=df['Country'],
                                 marker=dict(size=df['Population']/10,color=np.random.rand(5),showscale=True,opacity=0.8))])

fig.update_layout(title='Bubble Chart Showing Country Population vs GDP',xaxis_title='GDP',yaxis_title='Population',width=800,height=800)

fig.show()
```

## Q4. Histogram

```python
np.random.seed(0)
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
```

```python
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='skyblue', line_color='black')

show(p)
```
